### DATA HANDLING AND FEATURE ENGINEERING

```python
import re 
df['Round_lasted'] = df['verdict'].str.extract(r'round\s*(\d+)', flags=re.IGNORECASE)
df['Round_lasted'] = pd.to_numeric(df['Round_lasted'], errors='coerce')
df['Winner'] = df['verdict'].str.extract(r'^(.*?)\s+won', flags=re.IGNORECASE)

```

```python
df.drop('verdict',axis=1,inplace=True)
df.head()
df['Significant_Strikes_Landed'] = df['opponent_1_estimated_ability_to_take_punch'] - df['opponent_2_estimated_ability_to_take_punch']
df['weight_diff'] = df['opponent_1_avg_weight'] - df['opponent_2_avg_weight']
df['punch_power_diff'] = df['opponent_1_estimated_punch_power'] - df['opponent_2_estimated_punch_power']
df['punch_resistance_diff'] = df['opponent_1_estimated_punch_resistance'] - df['opponent_2_estimated_punch_resistance']
df['ko_percentage_diff'] = df['opponent_1_round_ko_percentage'] - df['opponent_2_round_ko_percentage']
df['opponent_1_won'] = (df['Winner']==df['opponent_1'].str.split().str[-1]).astype(int)
df['opponent_2_won'] = (df['Winner']==df['opponent_2'].str.split().str[-1]).astype(int)
df.head()
```

```python
df.drop('place',axis=1,inplace=True)
outcome_counts=df['Winner'].value_counts()
```

### MODEL IMPLEMENTATION

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier,StackingClassifier,VotingClassifier
import xgboost as xgb
import lightgbm as lgb
from sklearn.metrics import (
    accuracy_score, classification_report, 
    confusion_matrix, f1_score, roc_auc_score
)
```

```python
# defining features
feature_columns = [
    'opponent_1_estimated_punch_power', 'opponent_2_estimated_punch_power',
    'opponent_1_estimated_punch_resistance', 'opponent_2_estimated_punch_resistance',
    'opponent_1_estimated_ability_to_take_punch', 'opponent_2_estimated_ability_to_take_punch',
    'opponent_1_rounds_boxed', 'opponent_2_rounds_boxed',
    'opponent_1_round_ko_percentage', 'opponent_2_round_ko_percentage',
    'opponent_1_has_been_ko_percentage', 'opponent_2_has_been_ko_percentage',
    'opponent_1_avg_weight', 'opponent_2_avg_weight','Significant_Strikes_Landed',
         'weight_diff','punch_power_diff','punch_resistance_diff','ko_percentage_diff'
]
X=df[feature_columns]
y=df[['opponent_1_won',]]
```

```python
# Handling Missing Values
imputer=SimpleImputer(strategy='median')
X_imputed=imputer.fit_transform(X)
y_imputed=imputer.fit_transform(y)
```

```python
#Train_Test_Split
X_train,X_test,y_train,y_test=train_test_split(X_imputed,y_imputed,test_size=0.2,random_state=42,stratify=y)
```

```python
#Scaling Features
scaler = StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
models = {
    'LogisticRegression': LogisticRegression(random_state=42, max_iter=1000),
    'SVM': SVC(random_state=42, probability=True),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'AdaBoost': AdaBoostClassifier(random_state=42, n_estimators=100),
    'GradientBoosting': GradientBoostingClassifier(random_state=42, n_estimators=100),
    'XGBoost': xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
    'LightGBM': lgb.LGBMClassifier(random_state=42, n_estimators=100)
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Label encode target
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train.ravel() if hasattr(y_train, 'ravel') else y_train)
y_test_enc = le.transform(y_test.ravel() if hasattr(y_test, 'ravel') else y_test)

results = {}

for name, model in models.items():
    # Train the model
    model.fit(X_train_scaled, y_train_enc)
    
    # Predictions
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled) if hasattr(model, "predict_proba") else None
    
    # Metrics
    f1 = f1_score(y_test_enc, y_pred, average='weighted') if len(np.unique(y_train_enc)) > 2 else f1_score(y_test_enc, y_pred)
    accuracy = accuracy_score(y_test_enc, y_pred)
    
    # Store results
    results[name] = {
        'model': model,
        'accuracy': accuracy,
        'f1_score': f1,
        'predictions': y_pred,
        'probabilities': y_pred_proba,
        'classification_report': classification_report(y_test_enc, y_pred, output_dict=True)
    }
    
    # Print results
    print(f"\n{name} Results:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(classification_report(y_test_enc, y_pred))
```

```python
# Compare model performance
print("\n=== MODEL COMPARISON ===")
for name, result in results.items():
    print(f"{name}: Accuracy = {result['accuracy']:.4f}, F1 Score = {result['f1_score']:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import math

n_models = len(results)
n_cols = 3  # number of columns
n_rows = math.ceil(n_models / n_cols)  # number of rows

fig, axes = plt.subplots(n_rows, n_cols, figsize=(7*n_cols, 5*n_rows))

# Flatten axes array for easy indexing
axes = axes.flatten()

for i, (name, result) in enumerate(results.items()):
    cm = confusion_matrix(y_test_enc, result['predictions'])  # use encoded labels
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', ax=axes[i])
    axes[i].set_title(f'{name} Confusion Matrix')
    axes[i].set_xlabel('Predicted')
    axes[i].set_ylabel('Actual')

# Turn off any unused subplots
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python

#  Ensure target is 1D and encoded

y_train_flat = y_train.ravel() if hasattr(y_train, 'ravel') else y_train
y_test_flat = y_test.ravel() if hasattr(y_test, 'ravel') else y_test

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train_flat)
y_test_enc = le.transform(y_test_flat)


# Train & evaluate each model individually

results_eval = {}
valid_models_for_stack = {}

for name, model in all_models.items():
    print(f"\nTraining & Evaluating {name}...")
    try:
        model.fit(X_train_scaled, y_train_enc)
        y_pred = model.predict(X_test_scaled)
        
        # Metrics
        accuracy = accuracy_score(y_test_enc, y_pred)
        f1 = f1_score(y_test_enc, y_pred, 
                      average='weighted' if len(np.unique(y_train_enc)) > 2 else 'binary')
        report = classification_report(y_test_enc, y_pred)
        
        results_eval[name] = {
            'model': model,
            'accuracy': accuracy,
            'f1_score': f1,
            'predictions': y_pred,
            'classification_report': report
        }
        
        print(f"Accuracy: {accuracy:.4f}")
        print(f"F1 Score: {f1:.4f}")
        print(report)
        
        # Add model to stacking candidates
        valid_models_for_stack[name] = model
        
    except Exception as e:
        print(f"Skipping {name} due to error: {e}")


# 3 Optional: Voting Classifier
if len(valid_models_for_stack) >= 2:
    try:
        voting = VotingClassifier(
            estimators=[(name, model) for name, model in valid_models_for_stack.items()],
            voting='hard'  # safe if any model lacks predict_proba
        )
        voting.fit(X_train_scaled, y_train_enc)
        y_pred_vote = voting.predict(X_test_scaled)
        
        print("\nVoting Ensemble Results:")
        print("Accuracy:", accuracy_score(y_test_enc, y_pred_vote))
        print("F1 Score:", f1_score(y_test_enc, y_pred_vote,
                                     average='weighted' if len(np.unique(y_train_enc))>2 else 'binary'))
        print(classification_report(y_test_enc, y_pred_vote))
        
    except Exception as e:
        print(f"VotingClassifier failed: {e}")
else:
    print("Not enough models for voting ensemble (need at least 2).")


# 4 Optional: Stacking Classifier

if len(valid_models_for_stack) >= 2:
    try:
        stacking = StackingClassifier(
            estimators=[(name, model) for name, model in valid_models_for_stack.items()],
            final_estimator=RandomForestClassifier(random_state=42, n_estimators=200),
            cv=5,
            n_jobs=-1
        )
        stacking.fit(X_train_scaled, y_train_enc)
        y_pred_stack = stacking.predict(X_test_scaled)
        
        print("\nStacking Ensemble Results:")
        print("Accuracy:", accuracy_score(y_test_enc, y_pred_stack))
        print("F1 Score:", f1_score(y_test_enc, y_pred_stack,
                                     average='weighted' if len(np.unique(y_train_enc))>2 else 'binary'))
        print(classification_report(y_test_enc, y_pred_stack))
        
    except Exception as e:
        print(f"StackingClassifier failed: {e}")
else:
    print("Not enough models for stacking ensemble (need at least 2).")
```
